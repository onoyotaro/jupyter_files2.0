```python
import os
import sys
import glob
try:
    import numpy
except Exception:
    print("check modle; pip install numpy")  # モジュールの有無の警告
try:
    import pandas
except Exception:
    print("check modle; pip install pandas")


def main(args):
    """  サンプルファイル(IRデータを模したデータ)を生成して, 結合してExcelファイルで出力する
    # 例: 10個データを生成して, 結合します
    % python sample.py --num 10

    スクリプトを初めて実行する際に, sample_data を生成します.
    """
    if not os.path.isdir("./sample_data"):
        print("Make 'sample_data'...")
        os.mkdir("./sample_data")
    else:
        pass

    # 準備 :: サンプルファイルの生成
    # n倍角のsine 曲線のデータを 30個テキストファイルで作ります。
    x = numpy.arange(0, 360, 0.1)
    for i in range(1, args.num+1, 1):
        data = numpy.vstack((x, numpy.sin(numpy.radians(x*i)))).T
        numpy.savetxt(fname="sample_data/sine_%02d.txt" %
                      (i), X=data, delimiter=",", fmt="%1.5f")

    # 本題 :: データを連結して, Excel で出力します
    # -----------------------------------------
    txt_list = sorted(glob.glob("./sample_data/sine*"))  # ファイル名のリスト

    x_xls = numpy.loadtxt(
        txt_list[0], delimiter=",", usecols=1)  # 後述のExcel出力のための横軸

    for i, fname in enumerate(txt_list):
        _ = numpy.loadtxt(fname, delimiter=",",
                          dtype=numpy.float32,
                          usecols=1)  # csvファイル読み込み
        if i == 0:
            arr = _  # 最初のファイルのみ, arrと一致させる
        else:
            arr = numpy.vstack((arr, _))  # 2番目以降は, 直前の配列と結合する

    print(arr.shape)  # 要素数3600 行の30 列の配列。

    name = ["Sine: %d" % i for i in range(args.num)]  # ヘッダの作成

    # Excel ファイルで出力する為に, pandas.DataFrame 形式の変数を用意します。
    df = pandas.DataFrame(data=arr.T,  # データ
                          index=x_xls,  # 横軸; 実際の研究では波長とか周波数が入ります
                          columns=name,  # ヘッダ
                          )

    df.to_excel("./sample_data/templete.xlsx")  # Excel ファイルの出力


if __name__ in "__main__":
    import argparse

    p = argparse.ArgumentParser()
    p.add_argument("--num", default=10, type=int)

    args = p.parse_args()
    print(args)
    main(args)
```
